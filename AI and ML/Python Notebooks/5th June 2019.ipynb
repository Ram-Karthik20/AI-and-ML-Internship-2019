Handle missing data using pandas:
* dropna()
* fillna()
* interpolate()

```python
import pandas as pd
df=pd.read_csv('DataFiles/weather_data.csv')
df
```

```python
df.fillna(1.0) #It fills the value with the given numrical data
```

```python
df.fillna(method = "ffill") #ffill is a method called forward fill, which carry forwards the previous day's value
```

```python
df.fillna(method = "bfill") ##bfill is a method called backward fill, which carrys backward the next day's value
```

```python
df
```

```python
df.fillna(method ="ffill", axis = 'columns')  #axis=columns fills the NaN's vertically
```

#### Interpolate() is a function which comes up with a guess based on the other values in the dataset

```python
df.interpolate()
```

#### dropna() is a function which will drop rows or columns with missing values

```python
df.dropna()
```

```python
df1.dropna(how = "any") # how = any drops the rows where all the columns of the particular row are missing
```
